## 一、LLM 和 ChatModel

```python
# LLM 调用
from langchain.llms import OpenAI
import os


api_key=os.getenv("DASHSCOPE_API_KEY")
base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"

llm = OpenAI(
    model="qwen-plus",
    temperature=0,
    api_key=api_key, 
    base_url=base_url
)


llm.invoke("你好")





```

```python
# LLM 调用
from langchain_openai import ChatOpenAI
import os

api_key=os.getenv("DASHSCOPE_API_KEY")
base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"

llm = ChatOpenAI(
    model="qwen-plus",
    temperature=0,
    api_key=api_key, 
    base_url=base_url)

print(llm.invoke("你好"))

```
